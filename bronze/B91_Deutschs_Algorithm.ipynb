<h3> Quantum AND Gate </h3>

While describing the implementation of any boolean function $ f:\{0,1\}^n \rightarrow \{0,1\} $, we can use any of the existing universal gate sets. An example universal set would be using AND, OR and NOT gates. 

- An AND gate takes two inputs $x_1,x_2 \in \{0,1\}$ and returns 1 if and only if both $x_1$ and $x_2$ are equal to 1, 0 otherwise. 

- An OR gate takes two inputs $x_1, x_2 \in \{0, 1 \}$ and returns 1 if $x_1 = 1$ or $x_2 =1$.

- A NOT gate takes one input $x_1 \in \{0,1\}$ and returns the negation of $x_1$.

<h3> Deutsch's Problem </h3>

We say a boolean function $f:\{0,1\} \rightarrow \{0, 1\}$ is balanced if $f(0) \neq f(1)$. If $f(0) = f(1)$ we say that f is constant.

Deutch's problem is:
- Given f as an oracle, that is we can evaluate it for an input but we can't look inside, decide whether f is constant or balanced.

<h3>Classical Solution</h3>

Given such a function, we need to evaluate the function *twice* to get an answer using a classical computer. We had previously established that every 'classical' logical function $f$ can be converted to an equivalent unitary operator $U_f$ (by constructing a logical quantum circuit). Now we are going to propose a quantum algorithm that evaluates $U_f$ only *once*.

<h3> Solution </h3> 

We need to first look at what our oracle $U_f$ is doing to $\ket{0}$ and $\ket{1}$. (Findings from the previous task)
$$ U_f(\ket{x} \otimes \ket{a}) = \ket{x} \otimes \ket{a \oplus f(x)} $$
$$ U_f(\ket{x} \otimes \ket{0}) = \ket{x} \otimes \ket{0 \oplus f(x)} = \ket{x} \otimes \ket{\oplus f(x)} $$
$$ U_f(\ket{x} \otimes \ket{1}) = \ket{x} \otimes \ket{1 \oplus f(x)} $$
Now let us describe our proposed solution for this problem.

- Apply a X gate to the second qubit to get to our start state of $\ket{\psi_0} = \ket{0}\otimes \ket{1}$

- Apply Hadamard to both qubits. This will transform our state to $\ket{\psi_1} = \ket{+} \otimes \ket{-} $ (where $ \ket{+} = H\ket{0} $ and $ \ket{-} = H\ket{1} $)

- Let's write the expression as $\ket{\psi_1} = \frac{1}{\sqrt{2}}(\sum_{x\in\{0,1\}}\ket{x}\otimes\ket{-})$

- From the above expressions; we can rewrite $U_f$ to be $$ U_f(\ket{x} \otimes \frac{\ket{0}-\ket{1}}{\sqrt{2}}) = \frac{\ket{x}\otimes\ket{f(x)} - \ket{x}\otimes\ket{1\oplus f(x}}{\sqrt{2}} $$

- After we apply $U_f$ to $\psi_1$, our state will become: 

$$ \ket{\psi_2} = \frac{1}{\sqrt{2}}\bigg[ \sum_{x\in\{0,1\}} \frac{\ket{x}\otimes\ket{f(x)} - \ket{x}\otimes\ket{1\oplus f(x})}{\sqrt{2}} \bigg]$$

- So, if $f(x) = 0$, the expression in the summation is $ \ket{x}\otimes(\ket{0} - \ket{1}) $ and if $f(x) = 1$, it is $ \ket{x}\otimes(\ket{1} - \ket{0}) $. We can write the entire expression as $(-1)^{f(x)}\ket{x}\otimes(\ket{0} - \ket{1})$

- Now, after some rewriting, our entire state is 

$$ \ket{\psi_2} = \bigg[ \sum_{x\in\{0,1\}} \frac{1}{\sqrt{2}}(-1)^{f(x)}\ket{x}) \bigg] \otimes\frac{\ket{0} - \ket{1}}{\sqrt{2}}  $$

- If we focus on the first qubit (input) and ignore the second (output);

$$ \ket{\psi_{2,0}} = \sum_{x\in\{0,1\}} \frac{1}{\sqrt{2}}(-1)^{f(x)}\ket{x}) $$
$$ \ket{\psi_{2,0}} = \frac{1}{\sqrt{2}} \left[ (-1)^{f(0)}\ket{0} + (-1)^{f(x)} \right] $$
$$ \ket{\psi_{2,0}} = \frac{1}{\sqrt{2}}\myvector{(-1)^{f(0)} \\ (-1)^{f(1)}} $$

- When we apply $H$ gate to our qubit, we have the equation

$$ \ket{\psi_{3,0}} = \frac{1}{2}\mymatrix{rr}{1 & 1 \\ 1 & -1}\myvector{(-1)^{f(0)} \\ (-1)^{f(1)}} $$

- As an example, let's consider $ f(0) = f(1) = 1 $. 

$$ \ket{\psi_{3,0}} = \frac{1}{2}\mymatrix{rr}{1 & 1 \\ 1 & -1}\myvector{-1 \\ -1} = \myvector{-1 \\ 0} $$

- If we try out for the other cases; we get a $\pm\ket{0}$ when the function $f$ is constant and $\pm\ket{1}$ when it is balanced. So, we can find (with 100% certainty) whether $f$ is constant or balanced by a simple measurement of the first qubit.

Algorithm is as follows: 

- Apply a X gate to the second qubit.
- Apply Hadamard to both qubits.
- Apply $U_f$, that is call the oracle (*with your circuit*) 
- Apply Hadamard to first qubit.
- Measure first qubit and comment on the result.

<h5>Qiskit notes:</h5>

- You can use `circuit += oracle` to add the oracle to your whole circuit.
- It is possible to create a circuit with `QuantumCircuit(qubit_count, bit_count)` instead of going through the register route. In this case, you just pass the indices as arguments and not the qubit objects
- Barriers are not quantum programming primitives but they instruct qiskit to not apply any optimizations across the barrier. You can add them to your circuit using `circuit.barrier()`


```python
import random
from qiskit import QuantumCircuit, execute, Aer

# returns a 2 qubit circuit where qubit 0 is input and qubit 1 is output
def oracle():
    type = random.choice(["constant", "balanced"])
    result = QuantumCircuit(2)
    result.barrier()
    
    if type == "constant":
        # we ignore the input and randomly add a not gate
        if random.randrange(2) == 0:
            result.x(1)
    elif type == "balanced":
        # making sure the output is balanced
        result.cx(0, 1)
        # and randomly inverting it
        if random.randrange(2) == 0:
            result.x(1)
    
    result.barrier()
    return result

# dj_circuit = QuantumCircuit(2, 1)
# # build the circuit here ...
# dj_circuit.draw(output='mpl')

dj_circuit = QuantumCircuit(2, 1)
dj_circuit.x(1)

dj_circuit.h(0)
dj_circuit.h(1)

dj_circuit += oracle()

dj_circuit.h(0)
dj_circuit.measure(0, 0)

dj_circuit.draw(output='mpl')
```

```python
job = execute(dj_circuit, Aer.get_backend('qasm_simulator'),shots=10000)
counts = job.result().get_counts()
print(counts) # print the outcomes
```
